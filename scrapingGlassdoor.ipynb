```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd
```

```python
# Set the path to your ChromeDriver executable
webdriver_path = 'C:\Program Files\ChromeDriver\chromedriver.exe'

# Configure Chrome options
chrome_options = Options()
#chrome_options.add_argument(--headless)  # Run Chrome in headless mode
```

```python

base_url = 'https://www.glassdoor.fr/salaire/Dassault-Systemes-Salaires-E6162_P{}.htm'
# Define the total number of pages
total_pages = 3

# Create a new ChromeDriver instance
driver = webdriver.Chrome(service=Service(executable_path=webdriver_path), options=chrome_options)
datalist=[]
# Loop through the pages
for page_number in range(1, total_pages + 1):
    # Construct the URL for the current page
    url = base_url.format(page_number)
    # Navigate to the webpage
    driver.get(url)

    # Find the elements that contain the article titles
    element = driver.find_element(By.ID, 'SalariesRef') #By.CSS_SELECTOR, '#SalariesRef > div:nth-child(1) > div > div:nth-child(1) > div > a > strong')
    for i , child_element in enumerate (element.find_elements(By.XPATH, './*')):
        try:
            Nature=child_element.find_element(By.XPATH, 'div/div[1]/div/span').text.strip()
        except:
            Nature=""

        try:
            Bonus=int(child_element.find_element(By.XPATH, 'div/div[2]/div[2]/div[2]/div[2]/span[1]/strong/strong').text.replace('\u202f', '').replace('€', '')) 
        except:
            Bonus=0
        dic={'Position'   :child_element.find_element(By.XPATH, 'div/div[1]/div/a/strong').text.strip(),\
            'Nature'     :Nature,\
            'Salary'     :int(child_element.find_element(By.XPATH, 'div/div[2]/div[1]/span[1]/span/strong/strong').text.replace('\u202f', '').replace('€', '')) , \
            'Base Salary':int(child_element.find_element(By.XPATH, 'div/div[2]/div[2]/div[2]/div[1]/span[1]/strong/strong').text.replace('\u202f', '').replace('€', '')),
            'Bonus'      :Bonus }
        
        datalist.append(dic)
        driver.delete_all_cookies()
df=pd.DataFrame(datalist)
#driver.quit()
df.to_csv()
df
```
